**1.1 Dataset Overview**

The dataset used in this project is the California Housing Dataset, which contains information about housing districts in California. The data was originally derived from the 1990 U.S. Census and is widely used in machine learning research for housing price prediction tasks.

Each row in the dataset represents a housing district, and the features describe demographic, geographic, and housing-related characteristics of that district. These variables include geographic coordinates, population statistics, and housing supply indicators. The dataset also includes a categorical feature describing the district’s proximity to the ocean.

This dataset is appropriate for supervised machine learning, specifically a regression task, because the goal is to predict a continuous numeric variable representing housing value.

**Prediction Target**

The target variable (dependent variable) in this dataset is:

median_house_value

This variable represents the median house price in a given district, measured in U.S. dollars.

The machine learning task is therefore:

Predicting the median housing price of a district based on demographic, geographic, and housing characteristics.

**Why This Prediction Task Matters**

Predicting housing prices is an important real-world problem in real estate, urban planning, and financial analytics. Accurate housing price prediction models can support:

Home buyers in evaluating whether a property is reasonably priced

Real estate investors in identifying undervalued areas

Banks and mortgage lenders in assessing collateral value

Urban planners and policymakers in understanding how geographic and demographic factors influence housing affordability

From a machine learning perspective, this dataset is particularly valuable because it contains both numerical and categorical variables, allowing us to practice a full data science workflow including data preprocessing, feature engineering, predictive modeling, and explainability.

**Dataset Size and Feature Types**

The dataset contains:20,640 observations (rows),10 variables (columns)

Numerical Features (9)

These variables contain continuous numeric values:longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,median_house_value (target variable)

**Categorical Features (1)**

**ocean_proximity**

Describes the district's proximity to the ocean
Example categories include:NEAR BAY,<1H OCEAN,INLAND,NEAR OCEAN,ISLAND

This categorical variable will require encoding during the preprocessing stage before training machine learning models.

**Summary**

In summary, the California Housing dataset provides a rich set of demographic and geographic features for predicting district-level housing prices. With over twenty thousand observations and a mix of numerical and categorical features, the dataset is well-suited for exploring the full data science workflow, including exploratory analysis, regression modeling, and model explainability.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["MedHouseVal"], bins=50, kde=True)

plt.title("Distribution of Median House Value")
plt.xlabel("Median House Value")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,3))

sns.boxplot(x=df["MedHouseVal"])

plt.title("Boxplot of Median House Value")

plt.show()
```

**1.2 Target Distribution**

Target Variable: MedHouseVal (Median House Value)

The distribution of the target variable shows that housing prices are right-skewed. Most districts have median house values between approximately 1 and 3 (in units of $100,000), while fewer districts have very high housing prices.

There is a noticeable upper limit around 5, which indicates that the dataset has a capped value for house prices. This is a known characteristic of the California Housing dataset where the maximum value was truncated during data collection.

The distribution is therefore not perfectly normal and shows some skewness toward higher values. However, because this is a regression problem, class imbalance is not a concern in the same way as in classification tasks.

The presence of a price cap may slightly affect model performance for high-value districts. Tree-based models such as Random Forest and Gradient Boosting can help capture nonlinear relationships and handle this type of distribution effectively.

**1.3 Feature Distributions and Relationships**

Visualization 1: Median Income vs House Value

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x=df["MedInc"], y=df["MedHouseVal"], alpha=0.4)

plt.title("Median Income vs Median House Value")
plt.xlabel("Median Income")
plt.ylabel("Median House Value")

plt.show()
```

This scatter plot shows a strong positive relationship between median income and house prices. Districts with higher median income tend to have significantly higher housing values. However, the relationship is not perfectly linear, and we can observe a price cap near 5, suggesting that the dataset truncates very expensive housing values. This indicates that median income will likely be one of the most important predictive features in our models.

```python
# Visualization 2: Housing Age Distrubution by Price

plt.figure(figsize=(8,5))

sns.boxplot(x=pd.cut(df["MedHouseVal"], bins=5), y=df["HouseAge"])

plt.title("House Age Distribution by Housing Value")
plt.xlabel("House Value Bins")
plt.ylabel("Median House Age")

plt.show()
```

This boxplot shows how house age varies across different price levels. Older houses appear across all price categories, but higher-value districts tend to have a slightly wider distribution of housing ages. This suggests that location and economic conditions may influence house prices more strongly than the age of the houses themselves.

```python
# Visualization 3: Geographic Distribution of Housing Prices

plt.figure(figsize=(8,6))

scatter = plt.scatter(
    df["Longitude"],
    df["Latitude"],
    c=df["MedHouseVal"],
    cmap="viridis",
    alpha=0.5
)

plt.colorbar(scatter, label="Median House Value")

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographic Distribution of Housing Prices")

plt.show()
```

This geographic visualization reveals clear spatial patterns in housing prices across California. Higher housing values are concentrated near coastal regions, particularly around major metropolitan areas such as San Francisco and Los Angeles. Inland areas generally have lower housing prices. This indicates that location is a key determinant of housing value, and geographic features will likely play an important role in predictive modeling.

```python
# Visualization 4: Pairplot of Key Features
selected_features = ["MedInc", "HouseAge", "AveRooms", "Population", "MedHouseVal"]

sns.pairplot(df[selected_features], diag_kind="kde")

plt.show()
```

The pairplot shows relationships between several important numerical features and the target variable. Among these variables, median income shows the strongest positive relationship with housing prices. Other variables such as average rooms and population show weaker and more scattered relationships. This suggests that some features are much more predictive than others, which will influence feature importance in later modeling stages.

**1.4 Correlation Heatmap**

```python
# Compute correlation matrix
corr_matrix = df.corr()

# Display the matrix
corr_matrix
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap of Housing Features")

plt.show()
```

| Parameter       | Meaning                      |
| --------------- | ---------------------------- |
| annot=True      | shows correlation numbers    |
| cmap="coolwarm" | color scale for correlations |
| fmt=".2f"       | show values with 2 decimals  |
| linewidths      | adds separation lines        |


The correlation heatmap highlights the strength of relationships between different features in the dataset.

The strongest positive correlation with the target variable MedHouseVal is median income (MedInc). This indicates that districts with higher median income tend to have higher housing prices, suggesting that income is likely to be one of the most important predictors in our models.

Some variables such as AveRooms also show a moderate positive correlation with housing values, indicating that districts with larger homes tend to have higher prices. On the other hand, geographic variables such as Latitude and Longitude show weaker but noticeable correlations, suggesting that location plays a role in housing value differences.

There are also strong correlations among some independent variables. For example, total rooms, households, and population-related features are highly correlated, which may indicate multicollinearity. Tree-based models such as Random Forest and Gradient Boosting can handle correlated features better than linear models, which may influence model selection in later stages.
