**1.1 Dataset Overview**

The dataset used in this project is the California Housing Dataset, which contains information about housing districts in California. The data was originally derived from the 1990 U.S. Census and is widely used in machine learning research for housing price prediction tasks.

Each row in the dataset represents a housing district, and the features describe demographic, geographic, and housing-related characteristics of that district. These variables include geographic coordinates, population statistics, and housing supply indicators. The dataset also includes a categorical feature describing the district’s proximity to the ocean.

This dataset is appropriate for supervised machine learning, specifically a regression task, because the goal is to predict a continuous numeric variable representing housing value.

**Prediction Target**

The target variable (dependent variable) in this dataset is:

median_house_value

This variable represents the median house price in a given district, measured in U.S. dollars.

The machine learning task is therefore:

Predicting the median housing price of a district based on demographic, geographic, and housing characteristics.

**Why This Prediction Task Matters**

Predicting housing prices is an important real-world problem in real estate, urban planning, and financial analytics. Accurate housing price prediction models can support:

Home buyers in evaluating whether a property is reasonably priced

Real estate investors in identifying undervalued areas

Banks and mortgage lenders in assessing collateral value

Urban planners and policymakers in understanding how geographic and demographic factors influence housing affordability

From a machine learning perspective, this dataset is particularly valuable because it contains both numerical and categorical variables, allowing us to practice a full data science workflow including data preprocessing, feature engineering, predictive modeling, and explainability.

**Dataset Size and Feature Types**

The dataset contains:20,640 observations (rows),10 variables (columns)

Numerical Features (9)

These variables contain continuous numeric values:longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,median_house_value (target variable)

**Categorical Features (1)**

**ocean_proximity**

Describes the district's proximity to the ocean
Example categories include:NEAR BAY,<1H OCEAN,INLAND,NEAR OCEAN,ISLAND

This categorical variable will require encoding during the preprocessing stage before training machine learning models.

In summary, the California Housing dataset provides a rich set of demographic and geographic features for predicting district-level housing prices. **2.1 Data Preparation**

```python
# Define target variable
y = df["MedHouseVal"]

# Define features
X = df.drop("MedHouseVal", axis=1)

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

```python
print(X_train.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")

X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)
```

The dataset was prepared by separating the predictor variables (X) and the target variable (MedHouseVal). The data was then split into training (70%) and testing (30%) sets using a fixed random seed to ensure reproducibility. Missing values were handled using median imputation to prevent data loss and reduce the impact of outliers. Finally, feature scaling was applied using StandardScaler to normalize the numerical variables and improve model performance for algorithms sensitive to feature scale.

**2.2 Linear or Logistic Regression Baseline**

```python
from sklearn.linear_model import LinearRegression

# Initialize model
lr_model = LinearRegression()

# Train the model
lr_model.fit(X_train_scaled, y_train)
```

```python
# Predict on test data
y_pred_lr = lr_model.predict(X_test_scaled)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Compute metrics
mae = mean_absolute_error(y_test, y_pred_lr)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_lr))
r2 = r2_score(y_test, y_pred_lr)

print("Linear Regression Performance:")
print("MAE:", mae)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

plt.scatter(y_test, y_pred_lr, alpha=0.4)

plt.xlabel("Actual House Value")
plt.ylabel("Predicted House Value")
plt.title("Actual vs Predicted Housing Prices (Linear Regression)")

plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color="red")

plt.show()
```

The Linear Regression model serves as a baseline for predicting median housing prices. On the test dataset, the model achieved the reported MAE, RMSE, and R² values. The R² score indicates the proportion of variance in housing prices that can be explained by the model's features.

Although Linear Regression captures the general relationship between predictors and housing values, the scatter plot shows noticeable deviations between predicted and actual values. This suggests that the relationship between features and housing prices may be nonlinear, meaning more complex models such as Decision Trees, Random Forest, and Gradient Boosting may provide better predictive performance.

This baseline model provides a benchmark against which all subsequent models will be compared.

**2.3 Decision Tree/CART**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Define model
dt = DecisionTreeRegressor(random_state=42)

# Hyperparameter grid
param_grid = {
    "max_depth": [3, 5, 7, 10],
    "min_samples_leaf": [5, 10, 20, 50]
}

# GridSearchCV with 5-fold CV
grid_dt = GridSearchCV(
    estimator=dt,
    param_grid=param_grid,
    cv=5,
    scoring="neg_mean_squared_error",
    n_jobs=-1
)

# Fit on training data
grid_dt.fit(X_train_scaled, y_train)

print("Best hyperparameters:", grid_dt.best_params_)
print("Best CV score (neg MSE):", grid_dt.best_score_)
```

```python
best_dt = grid_dt.best_estimator_

# Predict on test set
y_pred_dt = best_dt.predict(X_test_scaled)

# Metrics
mae_dt = mean_absolute_error(y_test, y_pred_dt)
rmse_dt = np.sqrt(mean_squared_error(y_test, y_pred_dt))
r2_dt = r2_score(y_test, y_pred_dt)

print("Decision Tree Test Performance:")
print("MAE:", mae_dt)
print("RMSE:", rmse_dt)
print("R²:", r2_dt)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 8))
plot_tree(
    best_dt,
    feature_names=X.columns,
    filled=True,
    rounded=True,
    max_depth=3,   # show top levels for readability
    fontsize=9
)
plt.title("Decision Tree (Top 3 Levels Shown)")
plt.show()
```

```python
import joblib

joblib.dump(best_dt, "decision_tree_best.joblib")
joblib.dump(grid_dt.best_params_, "decision_tree_best_params.joblib")
```

The Decision Tree model was tuned using 5-fold cross-validation over max_depth and min_samples_leaf, optimizing for negative mean squared error. The best hyperparameters suggest a trade-off between model flexibility and overfitting control. On the test set, the Decision Tree improves interpretability by learning decision rules, but it may still underperform ensemble methods due to high variance. Tree visualization (top levels shown) provides a human-readable view of the most important splits driving predictions.

```python
dt_results = {
    "Model": "Decision Tree",
    "Best Params": grid_dt.best_params_,
    "MAE": mae_dt,
    "RMSE": rmse_dt,
    "R2": r2_dt
}
dt_results
```

**2.6 Neural Network**

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
input_dim = X_train_scaled.shape[1]

mlp_model = keras.Sequential([
    layers.Input(shape=(input_dim,)),
    layers.Dense(128, activation="relu"),
    layers.Dense(128, activation="relu"),
    layers.Dense(1, activation="linear")  # regression output
])

mlp_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss="mse",
    metrics=[keras.metrics.MeanAbsoluteError(name="mae")]
)

mlp_model.summary()
```

```python
history = mlp_model.fit(
    X_train_scaled,
    y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=64,
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(history.history["loss"], label="Train Loss (MSE)")
plt.plot(history.history["val_loss"], label="Val Loss (MSE)")
plt.title("MLP Training History: Loss")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(history.history["mae"], label="Train MAE")
plt.plot(history.history["val_mae"], label="Val MAE")
plt.title("MLP Training History: MAE")
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Predict
y_pred_mlp = mlp_model.predict(X_test_scaled).flatten()

# Metrics
mae_mlp = mean_absolute_error(y_test, y_pred_mlp)
rmse_mlp = np.sqrt(mean_squared_error(y_test, y_pred_mlp))
r2_mlp = r2_score(y_test, y_pred_mlp)

print("MLP Test Performance:")
print("MAE:", mae_mlp)
print("RMSE:", rmse_mlp)
print("R²:", r2_mlp)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred_mlp, alpha=0.4)
plt.xlabel("Actual House Value")
plt.ylabel("Predicted House Value")
plt.title("Actual vs Predicted (MLP)")

min_val = min(y_test.min(), y_pred_mlp.min())
max_val = max(y_test.max(), y_pred_mlp.max())
plt.plot([min_val, max_val], [min_val, max_val])

plt.show()
```

```python
mlp_model.save("mlp_regression.keras")
```

```python
mlp_results = {
    "Model": "MLP (Keras)",
    "Best Params": {"layers": [128, 128], "epochs": len(history.history["loss"]), "batch_size": 64},
    "MAE": mae_mlp,
    "RMSE": rmse_mlp,
    "R2": r2_mlp
}
mlp_results
```

The MLP model was trained with two hidden layers (128 units each) and ReLU activations using the Adam optimizer and MSE loss. The training and validation curves help assess whether the network is underfitting or overfitting; a widening gap between train and validation loss suggests overfitting. Test-set MAE, RMSE, and R² provide a direct comparison against linear regression and tree-based models. Because neural networks can capture nonlinear relationships, they may improve performance, but they also require careful regularization and tuning.

```python
configs = [
    {"units": 64, "lr": 0.001},
    {"units": 128, "lr": 0.001},
    {"units": 256, "lr": 0.001},
    {"units": 128, "lr": 0.0005},
]

results = []

for cfg in configs:
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(cfg["units"], activation="relu"),
        layers.Dense(cfg["units"], activation="relu"),
        layers.Dense(1, activation="linear")
    ])
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=cfg["lr"]),
                  loss="mse",
                  metrics=[keras.metrics.MeanAbsoluteError(name="mae")])

    h = model.fit(X_train_scaled, y_train, validation_split=0.2,
                  epochs=40, batch_size=64, verbose=0)

    pred = model.predict(X_test_scaled).flatten()
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    results.append((cfg, rmse))

results
```
